###Series - the 1D Structure


```python
marks=pd.Series([85,72,90,65])
print(marks)
print(type(marks))
```

```python
marks=pd.Series([85,72,90,65], index=["Ayesha", "Karim", "Sadia","Tamim"])
print(marks)
print(marks["Karim"])
print(type(marks))
```

```python
print(marks["Sadia"])
print(marks.values)
print(marks.index)
print(marks.dtype)
```

```python
price={
    "Rice":80,
    "Oil":180,
    "Egg":12
}
price_series=pd.Series(price)
print(price_series)
print(type(price))
```

###DataFrame - Creating Table

```python
data = {
    "name":["Ayesha","Karim","Sadia","Tamim"],
    "city":["Dhaka","Chattogram","Sylhet","Dhaka"],
    "marks":[85,72,90,65]
}

df=pd.DataFrame(data)
print(df)
```

```python
print(len(df))
```

```python
rows=[
    {"name":"Ayesha","city":"Dhaka","marks":85},
    {"name":"Karim","city":"Chattogram","marks":72}
]
df_another=pd.DataFrame(rows)
print(df_another)
```

```python
data2=[["Ayesha",85],["Karim",72]]
print(pd.DataFrame(data2,columns=["Name","Marks"]))
```

```python
data = {
    "name":["Ayesha","Karim","Sadia","Tamim"],
    "city":["Dhaka","Chattogram","Sylhet","Dhaka"],
    "marks":[85,72,90,65]
}
df=pd.DataFrame(data, index=["s1","s2","s3","s4"])
print(df)
```

```python
print(df.shape)
print(df.columns)
print(df.head(2))
print(df.tail(2))
```

```python
print(df.dtypes)
```

```python
print(df["name"])
print(type(df["name"]))
print(df[["name","marks"]])
```
